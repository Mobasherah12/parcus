```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
import torch
import os
import pandas as pd
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
import seaborn as sns
from pathlib import Path
import numpy as np
from torch.utils.data import DataLoader
from torch.utils.data.dataset import Subset, ConcatDataset
import torch.nn.functional as F


from datasets.NREDataset import NREHatespeechDataset, NRESpouseDataset
from datasets.utils import custom_collate
from training.NeuralPatternMatchingTraining import bagging
from training.utils import get_data_splits
```

```python
def colorize(words, opacity_array):
    # words is a list of words
    # color_array is an array of numbers between 0 and 1 of length equal to words
    string = ''
    for word, opacity in zip(words, opacity_array):
        string += f'<span style="background-color:hsl(360,100%,{int(opacity * 50 + 50)}%"> {word} </span>'
    return f'<div>{string}</div>'



def prepare_datasets(dataset_string, highlighted_path, non_highlighted_path, test_path):
    # We do not need a validation set here
    if dataset_string == 'NREHatespeechDataset':
        dataset_class = NREHatespeechDataset
    elif dataset_string == 'NRESpouseDataset':
        dataset_class = NRESpouseDataset
    else:
        raise

    highlighted_set = dataset_class(highlighted_path)
    non_highlighted_set = dataset_class(non_highlighted_path)

    all_train_dataset, train_split, validation_split, unlabelled_split = \
        get_data_splits(highlighted_set, non_highlighted_set, train_size, target_idx=3)

    all_train_dataset = ConcatDataset((highlighted_set, non_highlighted_set))

    train_set = Subset(all_train_dataset, train_split)
    test_set = dataset_class(test_path)

    return train_set, test_set


def get_token_info(model, dataset, idx=0, device='cpu'):

    _ = test_set[idx]  # allow caching of additional information
    tokens = test_set.cached[idx]['tokens']

    loader = DataLoader(Subset(dataset, [idx]), batch_size=1, collate_fn=custom_collate, shuffle=False)

    model.eval()

    for inputs, annotations, mask, targets, idx in loader:
        # print('New Batch (should have just one)')
        inputs, batch = inputs
        annotations, _ = annotations
        mask, _ = mask
        targets, _ = targets
         
        inputs = inputs.to(device)
        batch = batch.to(device)
        mask = mask.to(device)
        annotations = annotations.long().to(device)
        targets = targets.to(device)

        x = inputs
        
        # Run the forward pass.
        inputs = (inputs, annotations, mask, batch)
        classification, predictions = model(*inputs)
        
        
        deltas = model._compute_delta(x, model.prototypes)
        n_deltas = model._compute_not_delta(x, model.prototypes)
        
        
    return tokens, annotations, F.softmax(predictions, dim=1), F.softmax(classification, dim=1), deltas, n_deltas, targets


def generate_html(tokens, predictions):
    words = tokens
    opacity_array = predictions[:, 1].detach().numpy()
    s = colorize(words, 1-(opacity_array - 0.5))

    # to display in ipython notebook
    from IPython.display import display, HTML

    display(HTML(s))

    # or simply save in an html file and open in browser
    with open('colorize.html', 'w') as f:
        f.write(s)
```

```python
# Prepare dataset
model_string = 'NeuralPMSpouse'
dataset_string = 'NRESpouseDataset'
highlighted_train_path = str(Path(f'../data/Spouse/', f'Spouse_train/processed/highlighted'))
non_highlighted_train_path = str(Path(f'../data/Spouse/', f'Spouse_train/processed/'))
test_path = str(Path(f'../data/Spouse/', f'Spouse_test/processed/'))

train_size = 60
LFs = 1

train_set, test_set = prepare_datasets(dataset_string, highlighted_train_path, non_highlighted_train_path, test_path)

# Train models on a dataset
models = bagging(model_string, LFs, train_set, \
                 lr=1e-2, l1_coeff=1e-2, l2=1e-3, max_epochs=500,
                 no_prototypes=3, gating_param=100, batch_size=8,
                 save_path=None)
```

```python
# Get (token, prediction) for a sentence, for each model
# 8, 17 is a good idx (False positive)
for idx in range(100):
    o = get_token_info(models[0], test_set, idx=idx)
    tokens, annotations, predictions, classification, _, _, target = o
    #for t, a, p in zip(tokens, annotations, predictions):
    #    print(f'Token:{t} \t Ann: {a.detach()} \t Pred: {p.detach()} \t Out: {classification.detach()}, \t Target: {target.detach()}')

    # Generate html code for each model and store it into a file
    generate_html(tokens, predictions)
```

```python
import torch

model = models[0]
test_tokens = []
test_deltas = None
test_not_deltas = None
test_predictions = None

for idx in range(len(test_set)):
        
    o = get_token_info(model, test_set, idx=idx)
    tokens, annotations, predictions, classification, deltas, n_deltas, target = o
    
    test_tokens.extend(tokens)
    if test_predictions is None:
        test_predictions = predictions
    else:
        test_predictions = torch.cat((test_predictions, predictions), dim=0)

    if test_deltas is None:
        test_deltas = deltas
    else:
        test_deltas = torch.cat((test_deltas, deltas), dim=0)

    if test_not_deltas is None:
        test_not_deltas = n_deltas
    else:
        test_not_deltas = torch.cat((test_not_deltas, n_deltas), dim=0)

        
test_deltas = test_deltas.detach().numpy()
test_not_deltas = test_not_deltas.detach().numpy()
test_predictions = test_predictions.detach().numpy()
```

```python
tokens_not_deltas = list(zip(test_tokens.copy(), list(test_not_deltas)))
tokens_deltas = list(zip(test_tokens.copy(), list(test_deltas)))
tokens_predictions = list(zip(test_tokens.copy(), list(test_predictions[:, 1])))
```

```python
# SORT
# tokens_deltas = sorted(tokens_deltas, key=lambda token_delta: token_delta[1].max(), reverse=True)
# tokens_not_deltas = sorted(tokens_not_deltas, key=lambda token_not_delta: token_not_delta[1].max(), reverse=True)
tokens_predictions = sorted(tokens_predictions, key=lambda token_pred: token_pred[1], reverse=True)

token_pred = {}

for tp in tokens_predictions:
    if tp[0] in token_pred:
        token_pred[tp[0]].append(tp[1])
    else:
        token_pred[tp[0]] = [tp[1]]

token_avg_pred = {}
for k,v in token_pred.items():
    token_avg_pred[k] = sum(v)/len(v)

sorted_token_avg_pred = sorted(token_avg_pred.items(), key=lambda kv: kv[1], reverse=True)
```

```python
for tp in sorted_token_avg_pred[:100]:
    print(tp)
```

```python
no_words = 10
d = {'Word': [tp[0] for tp in sorted_token_avg_pred[:no_words]], 
     'Average Positive Score': [tp[1] for tp in sorted_token_avg_pred[:no_words]]
    }
df =pd.DataFrame(data=d)

plt.figure()#(figsize=(8, 6))
sns.set(font_scale=1.5)
sns.set_style('white')

ax = sns.barplot(x = 'Word', y = 'Average Positive Score', data=df, palette=sns.cubehelix_palette(1, start=2, rot=0, dark=0, light=.75, reverse=True))
for item in ax.get_xticklabels():
    item.set_rotation(90)
```
