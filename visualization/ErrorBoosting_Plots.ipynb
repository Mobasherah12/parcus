```python
import json
import os

import numpy as np
from pathlib import Path
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
os.environ['OMP_NUM_THREADS'] = "1"
```

```python
def parse_noise_results(results_folder):
    
    df = pd.DataFrame(columns=['Error Factor', 'Score', 'Training Size'])

    factors = [2.718281828459045, 5, 10]

    for factor in factors:

        for filename in [f for f in sorted(os.listdir(Path(results_folder + f'_ErrorFactor{factor}'))) if '.json' in f]:
            
            key = filename
            #print(key)

            with open(Path(Path(results_folder + f'_ErrorFactor{factor}'), filename), 'r') as f:
                result = json.load(f)
                
            if '10' in key:
                size = 10
            elif '300' in key:
                size = 300
            elif '30' in key:
                size = 30
            elif '60' in key:
                size = 60
            elif '150' in key:
                size = 150

            if factor > 2 and factor < 3:
                str_factor = 'e'
            else:
                str_factor = str(factor)
            df = df.append({'Error Factor': f'Factor {str_factor}', 'Score': float(result['test_scores']['f1']), 'Training Size': int(size)}, ignore_index=True)
                    
    return df
```

```python
if __name__ == '__main__':
    
    results_folder = '../results/Spouse/AdaptiveLFs/NeuralPMSpouse'
    df = parse_noise_results(results_folder)
    
    ax = sns.lineplot(hue="Error Factor", style="Error Factor", y="Score", x="Training Size", data=df)
    #plt.title('Impact of error factor ')
    plt.xticks([10, 30, 60, 150, 300])
    plt.tight_layout()
    plt.savefig('error_impact.png', dpi=600)
    plt.show()
```
